## Dimension Reduction Operations

For tensor shape intuition, these are very useful:

1. `sum(dim=...)` and `sum(-1)` (last dimension)
2. `mean(dim=...)`
3. `max(dim=...)` (returns both values and indices)
4. `keepdim=True` to preserve reduced axes as size `1`

`dim=-1` is especially common in deep learning because it targets the last axis regardless of batch dimensions.

```python
# 2D example: rows x cols = (2 x 3)
x2 = torch.tensor([[1, 3, 2],
                   [4, 0, 5]], dtype=torch.float32)
print("x2:\n", x2)
print("shape:", x2.shape)

print("=" * 20)
# Reduce last dimension (same as dim=1 for this 2D tensor)
print("sum(-1):", x2.sum(-1), " shape:", x2.sum(-1).shape)
print("sum(dim=1):", x2.sum(dim=1), " shape:", x2.sum(dim=1).shape)

print("=" * 20)
# Reduce first dimension
print("sum(dim=0):", x2.sum(dim=0), " shape:", x2.sum(dim=0).shape)

print("=" * 20)
# keepdim keeps reduced axis with size 1
print("sum(-1, keepdim=True):\n", x2.sum(-1, keepdim=True),
      " shape:", x2.sum(-1, keepdim=True).shape)

print("=" * 20)
# Other common reductions
print("mean(-1):", x2.mean(-1), " shape:", x2.mean(-1).shape)
max_vals, max_idx = x2.max(dim=-1)
print("max(-1) values:", max_vals, " indices:", max_idx)
```

### Shape-changing operations: reshape, view, squeeze, unsqueeze

These are core operations for preparing tensors in models.

- `reshape(...)`: returns a tensor with new shape (may copy if needed)
- `view(...)`: returns a reshaped **view** when memory is contiguous
- `squeeze(dim)`: removes axes of size 1
- `unsqueeze(dim)`: inserts a size-1 axis

```python
# reshape / view / squeeze / unsqueeze examples
x = torch.arange(12)
print("x:", x)
print("x shape:", x.shape)

print("=" * 20)
# reshape: general purpose
r = x.reshape(3, 4)
print("reshape(3,4):\n", r)
print("shape:", r.shape)

print("=" * 20)
# view: requires contiguous memory
v = x.view(3, 4)
print("view(3,4):\n", v)
print("shape:", v.shape)

print("=" * 20)
# Non-contiguous example: transpose breaks contiguity for view
t = r.t()  # shape (4, 3)
print("transposed shape:", t.shape, " contiguous:", t.is_contiguous())

try:
    t.view(12)
except RuntimeError as e:
    print("view on non-contiguous tensor failed:", e)

# reshape can still work (it may create a copy)
t_flat = t.reshape(12)
print("reshape on non-contiguous tensor works, shape:", t_flat.shape)

print("=" * 20)
# unsqueeze adds a size-1 dimension
u0 = x.unsqueeze(0)
u1 = x.unsqueeze(1)
print("unsqueeze(0) shape:", u0.shape)  # [1, 12]
print("unsqueeze(1) shape:", u1.shape)  # [12, 1]

print("=" * 20)
# squeeze removes size-1 dimensions
y = torch.randn(1, 12, 1)
print("y shape:", y.shape)
print("squeeze() shape:", y.squeeze().shape)      # [12]
print("squeeze(0) shape:", y.squeeze(0).shape)    # [12, 1]
print("squeeze(-1) shape:", y.squeeze(-1).shape)  # [1, 12]
```
